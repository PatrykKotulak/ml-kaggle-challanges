```python
import numpy as np
import pandas as pd

from functions.preprocessing import Imputer
```

```python
data = {
    'n1': [2, 2, 3, np.nan, 4, 2, 3, 1, np.nan, np.nan], # mean
    'n2': [2, 2, 3, np.nan, 4, 2, 3, 1, np.nan, np.nan], # median
    'n3': [2, 2, 3, np.nan, 4, 2, 3, 1, np.nan, np.nan], # most frequent
    'n4': [2, 2, 3, np.nan, 4, 2, 3, 1, np.nan, np.nan], # constant, fill_value = 10
    'n5': [2, 2, 3, -1, 4, 2, 3, 1, -1, -2],             # missing_values = -1
    's1': [np.nan, 'b', 'c', 'b', np.nan, 'e', np.nan, 'a', 'b', 'c'], # mean
    's2': [np.nan, 'b', 'c', 'b', np.nan, 'e', np.nan, 'a', 'b', 'c'], # median
    's3': [np.nan, 'b', 'c', 'b', np.nan, 'e', np.nan, 'a', 'b', 'c'], # most frequent
    's4': [np.nan, 'b', 'c', 'b', np.nan, 'e', np.nan, 'a', 'b', 'c'], # constant, fill_value = 'x'
    's5': [np.nan, 'b', 'c', 'b', np.nan, 'e', np.nan, 'a', 'b', 'c'], # add_indicator = True
    'g1': [1, 2, 1, 2, 1, 3, 2, 1, 0, 2]
}

data_new = {
    'n1': [2, np.nan, 3, 3, np.nan],
    's5': ['c', np.nan, 'c', 'c', np.nan]
}

options = {
    'n1': {'strategy': 'mean'},
    'n2': {'strategy': 'median'},
    'n3': {'strategy': 'most_frequent'},
    'n4': {'strategy': 'constant', 'fill_value': 10},
    'n5': {'strategy': 'mean', 'missing_values': -1},
    's1': {'strategy': 'constant', 'fill_value': 'fill'},
    's3': {'strategy': 'most_frequent'},
    's4': {'strategy': 'constant', 'fill_value': 'x'},
    's5': {'strategy': 'most_frequent', 'add_indicator': True},
}

df = pd.DataFrame(data=data)
df_new = pd.DataFrame(data=data_new)
```

```python
imp = Imputer(options=options)

imp.fit(df)
imp.save_options('options.json')

df_imputed = imp.transform(df)
df_imputed
```

```python
imp = Imputer.from_options_file('options.json')
imp.transform(df_new)
```
